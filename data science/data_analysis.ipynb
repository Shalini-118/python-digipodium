- selecting rows and columns 
 - rows (slicing)
   -`loc` - gets us the data using named indexes
   -`iloc`- gets us the data using numeric indexes
 - columns(dictionary style) 
   - `df(singles)` - requires the names of columns
   - `df(multiples)`- requires a list of columns (`multiple` is a list of named cods)

```python
# singles row(record)
weather.iloc[0]
```

```python
weather.iloc[-1]
```

```python
weather.iloc[25:40]
```

```python
# records 1-15 with first 5 col
weather.iloc[1:16,:5] # or [0,1,2,3,4] instead of:5
```
